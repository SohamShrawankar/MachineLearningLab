```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
```

```python
data = {
    'CustomerID': list(range(1,21)),
    'Gender': ['Male' ,'Male','Female','Female','Female','Female','Female','Female','Female','Female','Female','Female','Female','Female','Male','Male','Male','Male','Male','Male'],
    'Age':[19,21,20,23,31,22,35,23,64,30,67,35,58,24,37,22,35,20,52,35],
    'Annual Income (k$)': [15,15,16,16,17,17,18,18,19,19,19,19,20,20,20,20,21,21,23,23],
    'Spending Score (1-100)': [39,81,6,77,40,76,6,94,3,72,6,94,3,72,14,99,15,77,13,79]
}
```

```python
df = pd.DataFrame(data)
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i , init='k-means++' , random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
